#### Random `Integer` Generating  
- ```random.randint(num1, num2)```

````python
import random
random_int = random.randint(1,20)
print(random_int)
````

#### Random `Floating Point` numbers Generating  
- Generates random floating numbers between (0.0 to 1.0)   # but not exact 0.0 or 1.0  
- ```random.random()``` # float numbers generates between 0 and 1  
- 0.000000000000 to 0.99999999999

````python
import random
random_float = random.random() 
print(random_float)
````

##### How to generate a `random decimal number` between two numbers `(0,5)`
We cannot actually generate a float number bet any two numbers because it by default shows float values between 0 and 1 so we multiply with the highest number (5 in this case) and get the decimal random numbers between 0 and 5

````python
import random
random_decimal = random.random() * 5  #it will give you 0.0000000 to 4.9999999 not exactly zero(0) or five(5)
print(random_decimal)
````
