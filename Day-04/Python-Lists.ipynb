### **Lists**
([Click to Learn more about Lists](https://docs.python.org/3/tutorial/datastructures.html))
- are Data Structure in python
- Order in Data
- Order is determined by Order in the list such as [1, 2, 3, 5, 4, 7] # Output : 1, 2, 3, 5, 4, 7
- `print(var[index])` 
- like `print(var[0])` # Output: 1
- like `print(var[2])` # Output: 3
- like `print(var[-1])` # It will count from backwards. `[-1] is the last item on the lsit.` #Output : 7
#### **Data in the list can be modified too aka (Altering).**
- `var[1] = "8"`
so, after modifying the list should look like :[ 8, 2, 3, 5, 4, 7 ]
#### **Extend the existing list with another list.**
- Extented Data will be added at the end of the previous list.
- `var.extend( [9, 10] )` # Output : [ 8, 2, 3, 5, 4, 7, 9, 10 ]
#### **Data can be added too with .append() inside of the list.**
- Added Data will be added at the end of the list.
`var.append("Data")`

`````python
Divisions_of_the_BD = ["Dhaka", "Rajshahi", "Chittagong", "Rajshahi", "Khulna","Mymensingh"]
#print(Divisions_of_the_BD)
print(Divisions_of_the_BD[3])  # Output : Rajshahi
print(Divisions_of_the_BD[-1]) # Output : Mymensingh # [-1]2means the last item of the list.
print(Divisions_of_the_BD[-2]) # Output : Khulna # [-2] means the Second last item of the list.

#Modifying Existing Data
Divisions_of_the_BD[1] = "Sylhet"
print(Divisions_of_the_BD) #Output : ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna', 'Mymensingh']

#Add to the list
Divisions_of_the_BD.append("Rangpur") # Output: ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna', 'Mymensingh', 'Rangpur']
Divisions_of_the_BD.append("Rajshahi")
print(Divisions_of_the_BD) # Output: ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna', 'Mymensingh', 'Rangpur', 'Rajshahi']

#Extend the list with another list
Divisions_of_the_BD.extend(["Noakhali", "Kumilla"])
print(Divisions_of_the_BD)
# Output : ['Dhaka', 'Sylhet', 'Chittagong', 'Rajshahi', 'Khulna', 'Mymensingh', 'Rangpur', 'Rajshahi', 'Noakhali', 'Kumilla']
`````
