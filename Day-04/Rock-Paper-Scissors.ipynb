##### Rock Paper Scissors Game
- `Rock` wins against `Sissors`.
- `Scissors` wins against `Paper`.
- `Paper` wins agains `Rock`.

```python
import random
# Rock Paper Scissors ASCII Art
Rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''
Paper = '''
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
'''
Scissors= '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
""" print(Scissors) """
game_images = [Rock, Paper, Scissors]

user = int(input("What Do You Choose?\nType:  0 for `Rock` 1 for `Paper` 2 for `Scissors`\n"))

if user >=3 or user < 0:
    print("\n\nInvalid Choice!!!\nChoose between 0, 1 and 2\n")

else:
    
    print("\nYou Chose: ")
    print(game_images[user])

    computer_pick = random.randint(0, 2)

    print("\n\nComputer Chose: ")
    print(game_images[computer_pick])


    if user == 0 and computer_pick == 2:
        print("\nYou win!\n")

    elif computer_pick == 0 and user == 2: 
        print("\nYou lose!\n")
    elif computer_pick > user:
        print("\nYou lose!\n")
    elif user > computer_pick:
        print("\nYou win!\n")
    elif computer_pick == user: 
        print("\nIt's a Draw!\n")

```
