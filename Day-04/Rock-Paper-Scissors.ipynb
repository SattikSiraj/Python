##### Rock Paper Scissors Game
- `Rock` wins against `Sissors`.
- `Scissors` wins against `Paper`.
- `Paper` wins agains `Rock`.

```python
import random
# Rock Paper Scissors ASCII Art
Rock = '''Rock
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''
Paper = '''Paper
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
'''
Scissors= '''Scissors
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
""" print(Scissors) """
game_images = [Rock, Paper, Scissors]

user = int(input("What Do You Choose?\nType:  0 for `Rock` 1 for `Paper` and 2 for `Scissors`\n"))

if user >=3 or user < 0:
    print("\n\n\t\t\t\t\t\t    Invalid Choice!!!\n\n\t\t\t\t\t\tChoose between 0, 1 and 2\n")

else:
    
    #print("\nYou Chose: ")
    print(f"\nYou Chose: {game_images[user]}")

    computer_pick = random.randint(0, 2)

    #print("\nComputer Chose: ")
    print(f"\nComputer Chose: {game_images[computer_pick]}")

    if user == 0 and computer_pick == 2:
        print("\n\t\t\t\t\t\tYou win!\n")

    elif computer_pick == 0 and user == 2: 
        print("\n\t\t\t\t\t\tYou lose!\n")
    elif computer_pick > user:
        print("\n\t\t\t\t\t\tYou lose!\n")
    elif user > computer_pick:
        print("\n\t\t\t\t\t\tYou win!\n")
    elif computer_pick == user: 
        print("\n\t\t\t\t\t\tIt's a Draw!\n")



```
