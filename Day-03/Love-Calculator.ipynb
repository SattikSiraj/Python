var.lower() -> means all characters of the input variable will be `converted to lower case letters.`  
var.count("t") -> means it will `count` how many "t" are in the input variable. 

```python
print("The Love Calculator is calculating your score...\n")
name1 = input()
name2 = input()
combined_names = name1 + name2
lower_chars = combined_names.lower() # 
print(lower_chars)
t = lower_chars.count("t")
r = lower_chars.count("r")
u = lower_chars.count("u")
e = lower_chars.count("e")
first_digit = t + r + u + e
#print(first_digit)

l = lower_chars.count("l")
o = lower_chars.count("o")
v = lower_chars.count("v")
e = lower_chars.count("e")
second_digit = l + o + v + e
#print(second_digit)
score = int(str(first_digit) + str(second_digit))
#print(score)

if score<10 and score>= 90:
    print(f"The score is: {score}, you go together like coke and mentos!\n")
elif score>=40 and score <= 50:
    print(f"The score is: {score}, you are alright together.\n")
else:
    print(f"Your score is: {score}.\n")
```
