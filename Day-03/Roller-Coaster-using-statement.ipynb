# Conditional Statements

#### Operators with Meaning
- Greater than (>),
- Less than (<),
- Greater or equal (>=),
- Less than or equal (<=),
- Equal to (==),
- Not equal (!=)

#### Conditional Statement Structure.

```python
if condition:
    do this
elif condition:
    do this
else:
    DO THIS
```

````python
water_level = 81
if water_level > 80:
    print("Drain the Water")
else:
    print("Continue Pouring Water")
````

### Rollercoaster Ride

````python
print("Welcome to the Rollercoaster!\n")
height = int (input("What's is your height in cm\n"))

if height >= 120:
    print("You can ride the Rollercoaster!!")
else:
    print("Sorry! You cannot ride the Rollercoaster.")
````
