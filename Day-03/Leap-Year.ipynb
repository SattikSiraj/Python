### Leap Year 

- Step 1: if it is divisible by 4 then check if it is not divisible by 100 will be a `Leap.`
- Step 2: if it is divisible by 100 then check if it is divisible by 400
- Step 3: if it is divisible by 400 then it is a `Leap.`

![alt text](photo_2026-01-13_06-38-53.jpg)

```python
year = int(input("Enter year to check: \n"))

if year % 4 == 0:
    if year % 100 != 0:
        print(f"{year} is a Leap Year.\n")
    else:    
        if year % 400 == 0:
            print(f"{year} is a Leap Year.\n")
        else:
            print(f"{year} is not a Leap Year.\n")    
else:
    print(f"{year} is a Leap Year.\n")
```
