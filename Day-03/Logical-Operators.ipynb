```python
height = int(input("What's your height?\n"))
age = int(input("What is your age?\n"))
bill = 0
if height >= 120:
    print("You can ride!\n")
    if age<12:
        bill = 5
        print("Child tickets are:  $5\n")
    elif age <=18:
        bill = 7
        print("Youth tickets are:  $7\n")

        
    elif age >= 45 and age <= 55:
        bill = 0
        print(f"Everything is going to alright\nHave a free ride on us!\n")
    else:
        bill = 12
        print("Adult tickets are:  $12\n")

    wants_photo = input("Do you want a photo taken? Y or N\n")
    if wants_photo == "Y":
        print("For each photo you've to pay: $3\n")
        total_bill = bill + 3
        print(f"Final Bill is   ${bill} + $3 = ${total_bill}")
else:
    print("Sorry, you've to grow your height more than 120 cm\n")
```

```python
weight = int(input("Enter your Weight in kilogram: \n"))
height = float(input("Enter your Height in meters: \n"))
bmi = weight/ (height**2)
print(f"Your BMI is : {bmi:.2f}\n")

if bmi <= 18.5:
    print("You're Underweight.\n")
elif bmi>18.5 and bmi<=25:
    print("You have a Normal Weight.\n")
elif bmi>25 and bmi<=30:
    print("You're Slightly Overweight.\n")
elif bmi>30 and bmi<=35:
    print("You're in Obese.\n")
else:
    print("You're Clinically Obese.\n")
```
