#### Height Conversions
1 meter = 3.281 foot  
1 meter = 39.37 inch

```python
height_in_foot = int(input("Enter your height in Foot: \n"))
height_in_inch = int(input("Enter your height in inch: \n"))

total_height_in_meter = (height_in_foot/3.281) + (height_in_inch/39.37)

print(f"Your height in meters: {total_height_in_meter:.4f}")
```
