```python
height = float (input())
weight = int (input())
bmi = weight/(height**2)
print(round(bmi, 2))
```

- We use round(variable, 2) --> 2 values will be counted after the decimal point e.g.--> something.12

- print(round(2.666666666665,2)) output = 2.66

- print( 8 / 3 ) --> normal division --> output = 2.6666666665 
- print( 8 // 3 ) --> flow division --> output = 2
