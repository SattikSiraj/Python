# BMI = Weight / $Height^2$

![alt text](<Screenshot 2026-01-13 030341.png>)


```python
height = float (input())
weight = int (input())
bmi = weight/(height**2)
print(round(bmi, 2))
```

- We use round(variable, 2) --> 2 values will be counted after the decimal point e.g.--> something.12

- print(round(2.666666666665,2)) output = 2.66
