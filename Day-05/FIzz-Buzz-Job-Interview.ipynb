#### **FizzBuzz Game**  
- Your Program should print each number from 1 to a certain range in turn and include every numbers between the range.
- When the numbers is divisible by 3 then instead of printing the number it should print "Fizz"
- When the numbers is divisible by 5 then instead of printing the number it should print "Buzz"
- and if the is divisible by both 3 and 5 then instead of the number it should print "FizzBuzz"


**Key to solve this kind of problem is to approach it from reverse mentality of controlling the `if` statements**

```python
target = int(input())

for numbers in range(1, target+1):

    if numbers % 3 == 0 and numbers % 5 == 0:
        print("FizzBuzz")
    elif numbers % 3 == 0:
        print("Fizz")
    elif numbers % 5 ==0:
        print("Buzz")
    else:
        print(numbers)
```
