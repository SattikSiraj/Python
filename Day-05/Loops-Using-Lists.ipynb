### **Loops in Python**  
- A list, `list_of_items = ["Sattik", "Siraj", "Najmin", "Nahar"]`
- Syntax = *`for` item `in` list_of_items:*
- if you want to see an item `print(item)` # individual item of list.
- of you want to see the list `print(list_of_items)`
- A loop operation will continue until the list get empty.
- Loop will work counting every commands in the same indentation. (inside of a `for` loop) 


```python
fruits = ["Apple", "Peach", "Pear"]
count = 0

for fruit in fruits:
    print(fruit)
    print(fruit + " Pie")
    print(fruits) # indanted

#print(fruits)    # unindanted
```
