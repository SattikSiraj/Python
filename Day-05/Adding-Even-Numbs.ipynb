Take a input which determine the highest number and add all the even numbers between 1 and the `input`.

```python
target = int(input())
total = 0
for numbers in range(2, target+1, 2):
    total += numbers
print (total)
```

**Another Approach Can Be:**

```python
target = int(input())
total = 0
for numbers in range(2, target+1):
    if numbers % 2 == 0:
        total += numbers
print (total)
```
